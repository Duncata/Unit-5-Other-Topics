```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Establish site for scraping
url = 'https://en.wikipedia.org/wiki/List_of_2019_box_office_number-one_films_in_the_United_States'
html = urlopen(url)
```

```python
#Parse our html
soup = BeautifulSoup(html, 'lxml')
print(type(soup))
```

* The __th tag__ defines the header cells in the table which are displayed as bold, center-aligned text. ... 
* The __tr tag__ defines the table rows. There must be at least one row in the table. 
* The __td tag__ defines the standard cells in the table which are displayed as normal-weight, left-aligned text.

```python
table_rows = soup.find_all('tr')
print(table_rows[:10])
```

```python
trows = soup.find_all('tr')
print(trows[:5])
```

```python
#Print out each cell of each table row

for trow in trows:
    trow_td = trow.find_all('td')
print(trow_td)
type(trow_td)
```

```python
#Get text from each tr and td
list_trows = []

for trow in trows:
    cells = trow.find_all('td')
    str_cells = str(cells)
    cleantext = BeautifulSoup(str_cells, "lxml").get_text()
    list_trows.append(cleantext)
print(list_trows)
```

```python
#Create dataframe from output

df = pd.DataFrame(list_trows)
df.head(5)
```

```python
#split data on ',' 

df = df[0].str.split(',', expand=True)
df.head(10)
```

```python
# drop first row and first column
df = df.drop(0, axis=1)
df = df.drop(0, axis=0)
df.head()
```

```python
#split data on ' ' 

df_date = df[1].str.split(' ', expand=True)
df_date = df_date.iloc[:,[1,2]]
df_date.columns = ['month', 'day']
df_date.head(10)
```

```python
df = df.drop(1, axis=1)
df[['month', 'day']] = df_date[['month', 'day']]
#df = df_date + df
df.head()
```

```python
#
df['gross'] = df[df.columns[2:5]].apply(
    lambda x: ','.join(x.dropna().astype(str)), axis=1)
#
df = df.drop(df.columns[2:5], axis=1)
df.head()
```

```python
#Rename table headers as we accidentally parsed position into the team column
df.columns = df.columns = ['Year', 'Title','Note', 'None', 'None', 'None', 'None', 'None', 'None', 'Month', 'Day', 'Gross']
```

```python
#
df = df[:20]
```

```python
#
df['Gross'] = df['Gross'].str.strip('\n]')
df['Gross'] = df['Gross'].str.replace('$', '')
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = df['Gross'].astype('float', errors='ignore')
df['Gross']
```

```python
# Plot histograms for all continuous data
df['Gross'].hist()
plt.title("2019 Movie Gross")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```
